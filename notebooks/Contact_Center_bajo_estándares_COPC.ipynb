```python
#leer dataset
df = pd.read_csv('servical_contactcenter_RAW.csv')
```

```python
#contar nulos por columna
nulos = df.isnull().sum()
print(nulos[nulos > 0])
print((nulos / len(df) * 100).round(2)) #nulos en porcentaje
```

```python
#detecto duplicados
print(f"Duplicados exactos: {df.duplicated().sum()}")
```

```python
#ver cómo son los duplicados
df[df.duplicated(keep=False)].sort_values('id_llamada').head(15)
```

```python
#elimino duplicados (son exactamente iguales)
df = df.drop_duplicates()
df = df.reset_index(drop=True)

print(f"Filas después de eliminar duplicados: {len(df)}")
```

```python
#Elimino columna basura: columna_sistema_interno es constante, no aporta nada analítico
print(df['columna_sistema_interno'].nunique())
df = df.drop(columns=['columna_sistema_interno'])
print("Columna eliminada ✓")
```

```python
#limpio fechas con formato mixto
print(df['fecha'].head(10))
```

```python
#forzar parseo con dayfirst=True para el formato DD/MM/YYYY
df['fecha'] = pd.to_datetime(df['fecha'], dayfirst=True, errors='coerce')
# Verificar cuántas quedaron como NaT (no se pudieron parsear)
print(f"Fechas no parseadas: {df['fecha'].isna().sum()}")
```

```python
#Limpiar AHT_segundos (valores imposibles. uno válido es entre 30 segundos y 3600 segundos (1 hora)
print("Antes de limpiar AHT:")
print(df['AHT_segundos'].describe().round(2))
```

```python
# Ver los valores outliers AHT
print(df[df['AHT_segundos'] <= 0]['AHT_segundos'].value_counts())
print(df[df['AHT_segundos'] > 3600]['AHT_segundos'].value_counts())
```

```python
# Reemplazo con NaN los valores fuera de rango
df.loc[df['AHT_segundos'] <= 0, 'AHT_segundos'] = np.nan
df.loc[df['AHT_segundos'] > 3600, 'AHT_segundos'] = np.nan

print(f"\nNulos en AHT después de limpiar: {df['AHT_segundos'].isna().sum()}")
```

```python
#Limpiar tipo_gestion (variantes sucias) para luego calcular mediana según tipo de gestión e imputar valores nulos de AHT
print("Variantes únicas ANTES:")
print(sorted(df['tipo_gestion'].unique()))

# Limpiar espacios y estandarizar capitalización
df['tipo_gestion'] = df['tipo_gestion'].str.strip().str.title()

# Unificar plural → singular
df['tipo_gestion'] = df['tipo_gestion'].replace({
    'Consultas': 'Consulta',
    'Reclamos': 'Reclamo'
})

print("\nVariantes únicas DESPUÉS:")
print(sorted(df['tipo_gestion'].unique()))
```

```python
#Imputar AHT nulo con mediana por tipo de gestión
#¿Por qué mediana y no promedio? Porque es más resistente a outliers. y cada tipo de gestión tiene valores distintos, no es lo mismo un reclamo que una consulta

mediana_aht = df.groupby('tipo_gestion')['AHT_segundos'].transform('median')
df['AHT_segundos'] = df['AHT_segundos'].fillna(mediana_aht)

print("AHT imputado con mediana por tipo de gestión ✓")
print(df.groupby('tipo_gestion')['AHT_segundos'].median().round(0))
```

```python
# tratamiento de nulos en NPS_score
# ~10% de registros no tienen NPS porque el cliente cortó sin calificar la atención
# NO imputamos con la mediana global — imputamos por agente para ser más precisos

print(f"Nulos en NPS antes: {df['NPS_score'].isna().sum()}")

mediana_nps = df.groupby('id_agente')['NPS_score'].transform('median')
df['NPS_score'] = df['NPS_score'].fillna(mediana_nps)

# Si algún agente tiene todos nulos, usar mediana global como fallback
mediana_global_nps = df['NPS_score'].median()
df['NPS_score'] = df['NPS_score'].fillna(mediana_global_nps)

print(f"Nulos en NPS después: {df['NPS_score'].isna().sum()}")
```

```python
#Tratamiento de nulos en motivo_ausencia. Aclaración: estos nulos deberían ser VÁLIDOS cuando ausente = 0
# Solo hay problema cuando ausente = 1 y motivo_ausencia es NaN

ausentes_sin_motivo = df[(df['ausente'] == 1) & (df['motivo_ausencia'].isna())]
print(f"Ausentes sin motivo registrado: {len(ausentes_sin_motivo)}")

# Imputar como 'Sin datos' (es lo más real)
df.loc[(df['ausente'] == 1) & (df['motivo_ausencia'].isna()),
       'motivo_ausencia'] = 'Sin datos'

print("Imputado como 'Sin datos' ")
```

```python
#Limpio encoding roto en nombre_agente
print("Ejemplos con encoding roto:")
print(df[df['nombre_agente'].str.contains('Ã', na=False)]['nombre_agente'].unique())

# Reemplazar el nombre roto por el correcto
df['nombre_agente'] = df['nombre_agente'].str.replace('ValentÃ­na LÃ³pez',
                                                        'Valentina López',
                                                        regex=False)

print("Encoding corregido ")
```

```python
# Limpieza final y características listas para visualización de datos
#creo columnas nuevas que van a ser útiles en Power BI

# AHT en minutos (más legible en dashboard)
df['AHT_minutos'] = (df['AHT_segundos'] / 60).round(2)

# Rango de AHT (para segmentar en dashboard)
df['rango_AHT'] = pd.cut(df['AHT_segundos'],
                          bins=[0, 180, 300, 480, 9999],
                          labels=['Corta (<3min)', 'Normal (3-5min)',
                                  'Larga (5-8min)', 'Muy larga (>8min)'])

# NPS categorizado (Promotor / Neutro / Detractor)
def clasificar_nps(score):
    if score >= 9: return 'Promotor'
    elif score >= 7: return 'Neutro'
    else: return 'Detractor'

df['NPS_categoria'] = df['NPS_score'].apply(clasificar_nps)

# Flag de riesgo de rotación del agente
# Criterio: satisfaccion <= 2 Y ausente frecuente (proxy con ausente=1)
df['riesgo_agente'] = ((df['satisfaccion_agente'] <= 2) &
                        (df['ausente'] == 1)).astype(int)

# Mes y día de la semana (para análisis temporal)
df['mes'] = df['fecha'].dt.month
df['mes_nombre'] = df['fecha'].dt.strftime('%B')
df['dia_semana'] = df['fecha'].dt.day_name()
df['es_fin_de_semana'] = df['fecha'].dt.dayofweek.isin([5, 6]).astype(int)

print("Features creadas ✓")
print(df[['AHT_minutos', 'rango_AHT', 'NPS_categoria',
          'riesgo_agente', 'mes_nombre', 'dia_semana']].head())
```
